### Часть 1: MNIST
Обучите полносвязную нейронную сеть с архитектурой 784-100-100-10 и сигмоидой в качестве нелинейности. Какую точность классификации удалось получить? Нарисуйте график сходимости на обучающей и тестовой выборках. В качестве темпа обучения (learning rate) возьмите 0.01

```python
def predict(res):
    return np.argmax(res, axis=1)
```

```python
X_train, y_train = next(iter(train_loader_full))
X_train = X_train.view(X_train.shape[0], -1)
X_train = X_train.numpy()
y_train = y_train.numpy()

X_test, y_test = next(iter(test_loader_full))
X_test = X_test.view(X_test.shape[0], -1)
X_test = X_test.numpy()
y_test = y_test.numpy()
```

```python
def train(nn, loss_test, loss_train, accuracy_test, loss_train_b, epochs):
    for i in range(epochs):
        #print(i)
        res_test = nn.forward(X_test)
        loss_te = nll_loss.forward(res_test, y_test)
        loss_test.append(loss_te)

        res_train = nn.forward(X_train)
        loss_tr = nll_loss.forward(res_train, y_train)
        loss_train.append(loss_tr)

        accuracy = (predict(res_test) == y_test).sum()/len(res_test)
        accuracy_test.append(accuracy)

        for X, y in train_loader:
            X = X.view(X.shape[0], -1)
            X = X.numpy()
            y = y.numpy()
            res = nn.forward(X)
            loss_ = nll_loss.forward(res, y)
            loss_train_b.append(loss_)
            dLdy = nll_loss.backward()
            nn.backward(dLdy)
            nn.step(0.01)
```

```python
def plot_losses(loss_test, loss_train, accuracy_test, loss_train_b):
    plt.figure()
    figure(num=None, figsize=(20, 10), dpi=80, facecolor='w', edgecolor='k')
    plt.subplot(221)
    plt.title("Test loss")
    plt.plot(np.arange(len(loss_test)), loss_test)
    plt.subplot(222)
    plt.title("Train loss")
    plt.plot(np.arange(len(loss_train)), loss_train)
    plt.subplot(223)
    plt.title("Batch loss")
    plt.plot(np.arange(len(loss_train_b)), loss_train_b)
    plt.subplot(224)
    plt.title("Test accuracy")
    plt.plot(np.arange(len(accuracy_test)), accuracy_test)
    plt.show()
```

```python
nn = NeuralNetwork([Linear(784, 100), Sigmoid(), Linear(100, 100), Sigmoid(), Linear(100, 10)])
nll_loss = NLLLoss()
loss_test = []
loss_train = []
accuracy_test = []
loss_train_b = []
train(nn, loss_test, loss_train, accuracy_test, loss_train_b, 150)
```

```python
plot_losses(loss_test, loss_train, accuracy_test, loss_train_b)
```

```python
(predict(nn.forward(X_test)) == y_test).sum()/len(y_test)
```

Удалось достичь точности 0.9741. 

### Часть 2: Нелинейности

Обучите нейронную сеть с другими нелинейностями: ReLU и ELU. Сравните скорости сходимости и качество классификации с различными функциями активации.

```python
class SmoothReLU:
    def __init__(self):
        pass

    def forward(self, X):
        self.X = X
        self.Y = np.log(1 + np.exp(X))
        return self.Y
    
    def backward(self, dLdy):
        self.dLdx = dLdy * (1 / (1 + np.exp(-self.X)))
        return self.dLdx

    def step(self, learning_rate):
        pass
```

```python
class ReLU:
    def __init__(self):
        pass
    
    def forward(self, X):
        self.X = X
        self.Y = X
        self.Y[self.Y < 0] = 0
        return self.Y
    
    def backward(self, dLdy):
        self.dLdx = self.X
        self.dLdx[self.X <= 0] = 0
        self.dLdx[self.X > 0] = 1
        self.dLdx = dLdy * self.dLdx
        return self.dLdx
    
    def step(self, learning_rate):
        pass
```

```python
class ELU:
    def __init__(self):
        self.a = np.random.normal(0, 0.5, size=1)[0]
        if self.a < 0:
            self.a = 0
        pass
    
    def forward(self, X):
        self.X = X
        self.Y = self.X
        self.Y[self.X < 0] = self.a*(np.exp(self.Y[self.X < 0]) - 1)
        return self.Y
    
    def backward(self, dLdy):
        self.dLdx = self.X
        self.dLdx[self.X > 0] = 1
        self.dLdx[self.X <= 0] = self.a * np.exp(self.X[self.X <= 0])
        self.dLdx = self.dLdx * dLdy
        self.dLda = self.X
        self.dLda[self.X > 0] = 0
        self.dLda[self.X <= 0] = np.exp(self.X[self.X <= 0]) - 1
        self.dLda = np.sum(self.dLda * dLdy)
        return self.dLdx

    def step(self, learning_rate):
        self.a -= learning_rate * self.dLda  
        if(self.a < 0):
            self.a = 0
        pass
```

```python
nn_relu_s = NeuralNetwork([Linear(784, 100), SmoothReLU(), Linear(100, 100), SmoothReLU(), Linear(100, 10)])
loss_test_relu_s = []
loss_train_relu_s = []
accuracy_test_relu_s = []
loss_train_b_relu_s = []

train(nn_relu_s, loss_test_relu_s, loss_train_relu_s, accuracy_test_relu_s, loss_train_b_relu_s, 150)
```

```python
plot_losses(loss_test_relu_s, loss_train_relu_s, accuracy_test_relu_s, loss_train_b_relu_s)
```

```python
nn_relu = NeuralNetwork([Linear(784, 100), ReLU(), Linear(100, 100), ReLU(), Linear(100, 10)])

loss_test_relu = []
loss_train_relu = []
accuracy_test_relu = []
loss_train_b_relu = []

train(nn_relu, loss_test_relu, loss_train_relu, accuracy_test_relu, loss_train_b_relu, 150)
```

### HW1 (1 балл): Реализовать Tanh

```python
class Tanh:
    def __init__(self):
        pass
    
    def forward(self, X):
        self.X = X
        self.Y = np.tanh(X)
        return self.Y
    
    def backward(self, dLdy):
        self.dLdx = 1 - self.Y * self.Y
        self.dLdx *= dLdy
        return self.dLdx
    
    def step(self, learning_rate):
        pass
```

Какой прирост дают эти аугментации вместе и по отдельности

```python
class AddNoise(object):
    def __init__(self, mean=0., std=0.2):
        self.std = std
        self.mean = mean
        
    def __call__(self, tensor):
        tensor += torch.from_numpy(self.mean + self.std*
                                   np.random.normal(size=tensor.numpy().shape))
        return tensor
    
    def __repr__(self):
        return self.__class__.__name__ + '(mean={0}, std={1})'.format(self.mean, self.std)
```
